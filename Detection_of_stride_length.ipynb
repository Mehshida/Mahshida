```python
import numpy as np
import scipy.signal as signal

# Example function to detect heel strikes and toe offs
def detect_strides(marker_data, threshold=0.1, smoothing_window=5):
    # Apply a moving average filter for smoothing
    smoothed_data = np.convolve(marker_data, np.ones(smoothing_window)/smoothing_window, mode='valid')

    # Calculate the vertical displacement or velocity
    displacement = np.diff(smoothed_data)

    # Detect peaks and troughs in displacement to find heel strikes and toe offs
    peaks, _ = signal.find_peaks(displacement, height=threshold)
    troughs, _ = signal.find_peaks(-displacement, height=threshold)

    # Combine and sort peaks and troughs
    events = np.sort(np.concatenate((peaks, troughs)))

    # Return detected events
    return events

# Example usage
marker_data = np.array([0, 1, 2, 1, 0, -1, -2, -1, 0, 1, 2, 1, 0])  # Example marker data
stride_events = detect_strides(marker_data)

print("Detected stride events:", stride_events)
```
