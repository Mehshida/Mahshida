# K-Means Clustering

```
import  numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values # The iloc function is used to select rows and columns by their integer index. we use x and y because x is y is outtup we have to find
```

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
   kmeans = KMeans(n_clusters = i, init = "k-means++", random_state = 42 )
   kmeans.fit(X)
   wcss.append(kmeans.inertia_)
plt.plot( range(1, 11), wcss)
plt.title("the elbow method")
plt.xlabel("no of clusters")
plt.xlabel("wcss")
plt.show ()
```

```
kmeans = KMeans(n_clusters = 5, init = "k-means++", random_state = 42 )
y_kmeans = kmeans.fit_predict(X)
```

```
print(y_kmeans)
```

```
plt.scatter(X[y_kmeans == 0, 0] , X[y_kmeans == 0, 1],s = 100, c ='red', label='cluster1')
plt.scatter(X[y_kmeans == 1, 0] , X[y_kmeans == 1, 1],s = 100, c ='blue', label='cluster1')
```
